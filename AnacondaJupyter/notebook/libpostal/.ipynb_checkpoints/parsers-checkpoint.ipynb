```python
def parseWithPelias(query=''):
    ms1 = time.time_ns()
    #url = f"http://peliasparser.shared.madiva.vpn/parser/parse?text=" + query + "&debug=false"
    url = f"http://192.168.0.17:6800/parser/parse?text=" + query + "&debug=false"
    r = requests.get(url)
    ms2 = time.time_ns()
    res={
        "engine": "pelias",
        "ms": (ms2-ms1)/1000000
    }
    #print("R:" + str(r.json()));
    found={}
    result=r.json()
    
    print(json.dumps(result, indent=4))
    
    if (len(result["solutions"])>0):
        for ele in result["solutions"][0]["classifications"]:
            if(ele["label"]=="street"):
                found["via"]=ele["value"]
            elif(ele["label"]=="housenumber"):
                found["num"]=ele["value"]
            elif(ele["label"]=="locality"): 
                found["localidad"]=ele["value"]
            elif(ele["label"]=="region"): 
                found["region"]=ele["value"]                
            elif(ele["label"]=="country"): 
                found["pais"]=ele["value"]                                
            else:
                print("Pelias: No se que es esto:" + str(ele))

    if(not found.__contains__("pais")):found["pais"]="-"   
    if(not found.__contains__("region")):found["region"]="-"
    if(not found.__contains__("localidad")):found["localidad"]="-"
    if(not found.__contains__("cp")):found["cp"]="-"   
    if(not found.__contains__("via")):found["via"]="-"
    if(not found.__contains__("num")):found["num"]="-"
    if(not found.__contains__("esc")):found["esc"]="-"
    if(not found.__contains__("piso")):found["piso"]="-"
    if(not found.__contains__("puerta")):found["puerta"]="-"
        
    res["result"]=found
    return res

#direccion = "caleruega 18 4ºB madrid madrid"
direccion = "West 26th Street, New York, NYC, 10010"
pelias = parseWithPelias(direccion.upper())
print(json.dumps(pelias, indent=4))
```

```python
def analizaDirecciones(direcciones, engine='libpostal'):
    df=pd.DataFrame()
    for d in direcciones:
        d=d.upper()
        if (engine=='libpostal'):
            parsed = parseWithLibpostal(d)
        elif(engine=='pelias'):
            parsed = parseWithPelias(d)
        else:
            print("Error: Unknown engine " + engine)
            return None
        #print("libpostal:" + str(parsed))
        d1 = {'direccion': d, 'pelias_ms': parsed["ms"],   
              'via': parsed["result"]["via"],   'num': parsed["result"]["num"],   
              'ciudad': parsed["result"]["region"], 'region': parsed["result"]["region"],
             'cp': parsed["result"]["cp"], 'pais': parsed["result"]["pais"],
              'esc': parsed["result"]["esc"], 'piso': parsed["result"]["piso"],
              'puerta': parsed["result"]["puerta"]
             }
        #df1 = pd.DataFrame(d1, index={1})
        df = pd.concat([df], axis=0).reset_index(drop=True)

    return df
```

```python
def buscaDiccionarios(palabra):
    if(len(palabra)==0): return []
    lista=["cp"]
    if(diccionarioProvincias.check(palabra)):lista.append("region")
    if(diccionarioMunicipios.check(palabra)):lista.append("ciudad")
    if(diccionarioVias.check(palabra)):lista.append("via")
    if(any(elem in palabra for elem in r"ºªoa")):lista.append("piso")
    if(palabra=="ESPAÑA"):lista.append("pais")
    return lista
```

```python
def masParecido(palabra, lista):
    dict,max = {},0
    for l in lista:
        tmp = difflib.SequenceMatcher(None, palabra, l).ratio();
        dict[tmp] = l
        if tmp > max:
           max = tmp

    return(dict[max], max)

#masParecido("prfomnc", ['prominence', 'performance', 'preform', 'Provence', 'preferment', 'proforma'])
```

```python
def buscaParecidos(palabra):
    if(len(palabra)<=2): return []
    lista=[]
    
    #provParecidas=diccionarioProvincias.suggest(palabra)
    provParecida=masParecido(palabra, diccionarioProvincias.suggest(palabra))
    if (len(provParecida)>0): 
        t = (provParecida, "region")
        lista.append(t)

    muniParecido=masParecido(palabra, diccionarioMunicipios.suggest(palabra))
    if (len(muniParecido)>0): 
        t = (muniParecido, "ciudad")
        lista.append(t)

    viaParecido=masParecido(palabra, diccionarioVias.suggest(palabra))
    if (len(viaParecido)>0): 
        t = (viaParecido, "via")
        lista.append(t)
    
    
    return lista

buscaParecidos('MALAGO')
```

```python
def validarCandidatos(palabra, candidatos):
    candidatosValidados = set()
    cand=candidatos.copy()
    regexCP = '^[0-9]?[1-9][0-9][0-9][0-9]$'
    regexSololetras = '^[A-Za-z]+$'
    for c in candidatos:
        #print("Evaluando " + palabra + " como candidato a " + c)
        if (c=="cp"):
            if (re.match(regexCP, palabra)): 
                #print(palabra + " puede ser un CP")
                candidatosValidados.add(c)
            else:
                candidatosValidados.add(cand.remove("cp"))
        elif(c=='pais' or c=='region' or c=='ciudad' or c=='via'):
            if (re.match(regexSololetras, palabra)): 
                #print(palabra + " puede ser " + c)
                candidatosValidados.add(c)
        else:
            print("No se como validar el campo " + c)

    if (candidatosValidados.__contains__(None)): candidatosValidados.remove(None)
    return candidatosValidados
        
#s={"cp","ciudad","region", "pais"}        
#r=validarCandidatos("sss", s)
#print("R:" + str(r))
```

```python
def refinarVia(d):
    numbersFound = re.findall('[0-9]+', d["df"].loc[index,"via"])
    if(len(numbersFound) == 1):
        print("\t\tEncontrado un numero en la via, descomponiendo")
        d["df"].loc[index,"via"]= d["df"].loc[index,"via"].replace(numbersFound[0], '').strip() 
        if(d["df"].loc[index,"num"]!="-"):
            if(d["df"].loc[index,"piso"]=="-"):
                print("\t\tNumero ya tiene "+d["df"].loc[index,"num"]+", moviendolo a piso porque esta vacio")
                d["df"].loc[index,"piso"]=d["df"].loc[index,"num"]
                d["df"].loc[index,"num"]=numbersFound[0].strip()
            else: 
                print("\t\tNumero ya tiene '" + d["df"].loc[index,"num"] + "', no puedo moverlo a piso porque piso ya tiene '" + d["df"].loc[index,"piso"] + "'")
    
    elif(len(numbersFound) >= 2):
        print("\tEncontrados varios numeros en la via, descomponiendo")
        separadas=d["df"].loc[index,"via"].split(numbersFound[0])
        separadas = list(map(str.strip, separadas))
        
        if(len(numbersFound)==1 or len(numbersFound)==2):
            d["df"].loc[index,"via"]= separadas[0]
            d["df"].loc[index,"num"]= numbersFound[0]
            d["direccionNoIdentificada"]=d["direccionNoIdentificada"].replace(separadas[0], '')
            d["direccionNoIdentificada"]=d["direccionNoIdentificada"].replace(numbersFound[0], '')
            d["identificados"].add('num')
            if ('num' in d["noIdentificados"]): d["noIdentificados"].remove('num')
            
            if(len(numbersFound)==2):
                moreNumbersFound = re.findall('[0-9]+', separadas[1])
                if(len(moreNumbersFound)==1):
                    d["df"].loc[index,"piso"]= moreNumbersFound[0]
                    d["identificados"].add('piso')
                    d["noIdentificados"].remove('piso')

                    posibleLetra=separadas[1].replace(moreNumbersFound[0],'')
                    print("\t\tPosible Letra:" + str(posibleLetra))
                    if(len(posibleLetra)<=2 and any(elem in posibleLetra for elem in r"ºªoa")):
                        d["df"].loc[index,"puerta"]= re.sub(r'[ºªoa]?', '', posibleLetra)
                        d["identificados"].add('puerta')
                        d["noIdentificados"].remove('puerta')
        else:
            print("Hay demasiados numeros en la via, no se")
```

```python
def refinarNumero(d):
    print("\t\tRefinando numero '" + d["df"].loc[index,"num"] + "'")
    moreNumbersFound = re.findall('[0-9]+', d["df"].loc[index,"num"])
    if(len(moreNumbersFound)==1 and 'piso' not in d["identificados"] and 'puerta' not in d["identificados"]):
        posiblePisoLetra=d["df"].loc[index,"num"]
        if(re.match('^[0-9][A-Z]$', posiblePisoLetra)):
            print("moreNumbersFound:" + str(moreNumbersFound))
            d["df"].loc[index,"piso"]= moreNumbersFound[0]
            d["df"].loc[index,"puerta"]= re.sub(r'[ºªoa]?', '', posiblePisoLetra).replace(moreNumbersFound[0], '')
            identificados.add('piso')
            noIdentificados.remove('piso')
            identificados.add('puerta')
            noIdentificados.remove('puerta')  
    elif(len(moreNumbersFound)==2 and 'piso' not in d["identificados"] and 'puerta' not in d["identificados"]):
        separadas=d["df"].loc[index,"num"].split(moreNumbersFound[0])
        d["df"].loc[index,"num"]= moreNumbersFound[0] 
        posiblePisoLetra=separadas[1].replace(moreNumbersFound[0],'').strip()
        if(len(posiblePisoLetra)<=2 and (re.match('^[0-9][A-Z]$', posiblePisoLetra)  or any(elem in posiblePisoLetra for elem in r"ºªoa"))):
            d["df"].loc[index,"piso"]= moreNumbersFound[1] 
            d["df"].loc[index,"puerta"]= re.sub(r'[ºªoa]?', '', posiblePisoLetra).replace(moreNumbersFound[1], '') 
            d["identificados"].add('piso')
            d["noIdentificados"].remove('piso')
            d["identificados"].add('puerta')
            d["noIdentificados"].remove('puerta') 
```

```python
def refinarPiso(d):
    print("\t\tRefinando piso '" + d["df"].loc[index,"piso"] + "'")
    if(d["df"].loc[index,"piso"]!="-" and d["df"].loc[index,"puerta"]=="-" and re.match('^[0-9][ºªoa]?[A-Z]$', d["df"].loc[index,"piso"])):
        moreNumbersFound = re.findall('[0-9]+', d["df"].loc[index,"piso"])
        if(len(moreNumbersFound)==1):
            separadas=d["df"].loc[index,"piso"].split(moreNumbersFound[0])
            posibleLetra=separadas[1].replace(moreNumbersFound[0],'').strip()
            print("\t\tEncontrado solo un numero, posible letra:" + posibleLetra)
            if(re.match('^[ºªoa]?[A-Z]$', posibleLetra)):
                d["df"].loc[index,"piso"]= moreNumbersFound[0] 
                d["df"].loc[index,"puerta"]= re.sub(r'[ºªoa]?', '', posibleLetra) 
                d["identificados"].add('piso')
                d["noIdentificados"].remove('piso')
                d["identificados"].add('puerta')
                d["noIdentificados"].remove('puerta')  
```
